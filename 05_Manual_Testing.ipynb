```python
import joblib
import pandas as pd
```

```python
rf = joblib.load("../models/fraud_detection_model.pkl")
scaler = joblib.load("../models/amount_scaler.pkl")
```

```python
data = pd.read_csv("../data/creditcard.csv")

X = data.drop("Class", axis=1)
y = data["Class"]

X["Amount"] = scaler.transform(X[["Amount"]])
```

```python
single_transaction = X.iloc[100].values.reshape(1, -1)

prediction = rf.predict(single_transaction)
probability = rf.predict_proba(single_transaction)[0][1]

print("Prediction:", "Fraud" if prediction[0] == 1 else "Genuine")
print("Fraud Probability:", probability)
```

```python
rf.predict(X.iloc[0:10])
```

```python
# threshold based testing

threshold = -0.2
custom_prediction = (probability > threshold)

print("Custom Threshold Prediction:", "Fraud" if custom_prediction else "Genuine")
```
